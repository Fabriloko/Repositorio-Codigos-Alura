## Removendo os imoveis comerciais

```python
print(dados.Tipo.unique())
imoveis_comerciais = ['Conjunto Comercial/Sala',
                      'Prédio Inteiro', 'Loja/Salão',
                      'Galpão/Depósito/Armazém',
                      'Casa Comercial', 'Terreno Padrão',
                      'Loja Shopping/ Ct Comercial',
                      'Box/Garagem', 'Chácara',
                      'Loteamento/Condomínio', 'Sítio',
                      'Pousada/Chalé', 'Hotel', 'Indústria',
                      ]
```

```python
df_imoveis_residenciais = dados.query('@imoveis_comerciais not in Tipo')
df_imoveis_residenciais.Tipo.unique()
```

```python
df_preco_tipo = df_imoveis_residenciais.groupby('Tipo')[['Valor']].mean().sort_values('Valor')
df_preco_tipo.plot(kind= 'barh', figsize= (14,10))
```

```python
df_apartamentos = df_imoveis_residenciais.query("Tipo == 'Apartamento'").copy()
```

## Tratando Dados Nulos

```python
df_apartamentos.isnull().sum()
```

```python
df_apartamentos.fillna(0, inplace= True) #Utilizado para substituir os dados nulos
# method= "ffill" -> Metodo de preenchimento para dados nulos anterior
# method= "bfill" -> Metodos de preenchimento para dados nulos posterior 
# df_apartamentos.interpolate() #Utilizado para interpolar os dados nulos
```

```python
registros_a_remover = df_apartamentos.query('Valor == 0 | Condominio == 0').index
df_apartamentos.drop(registros_a_remover, inplace= True)
```

```python
df_apartamentos.Tipo.unique() # Coluna Tipo só tem o valor 'Apartamento'
```

```python
df_apartamentos.drop('Tipo', axis= 1, inplace= True)
```

## Filtrando o DataFrame

```python
#Filtrando Quartos e Valor
df_apartamentos_1 = df_apartamentos[(df_apartamentos['Quartos'] == 1) & 
                                    (df_apartamentos['Valor'] < 1200)]
```

```python
#Filtrando Quartos ,Valor e Area
df_apartamentos_2 = df_apartamentos[(df_apartamentos['Quartos'] >= 2) & 
                                    (df_apartamentos['Valor'] < 3000) &
                                    (df_apartamentos['Area'] > 70)]
```

## Criando colunas Numericas

```python
dados['Valor_por_mes'] =   dados['Valor'] + dados['Condominio']
dados.head()
```

```python
dados['Valor_por_ano'] = dados['Valor_por_mes'] *12 + dados['IPTU']
dados.head()
```

## Criando colunas Categoricas

```python
dados['Descicao'] = dados['Tipo'] + ' com ' + dados['Quartos'].astype(str) + \
                    ' quarto(s) e ' + dados['Area'].astype(str) + ' m2' + \
                    ' localizado no ' + dados['Bairro'].astype(str) + ' com ' + \
                    dados['Vagas'].astype(str) + ' vaga(s) de garagem'
dados.head()
```

```python
dados['Possui_suite'] = dados['Suites'].apply(lambda x: "Sim" if x > 0 else 'Não')
dados.head()
```
